```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('A:\\Data Analysis Jupyter\\Pollution\\assam_pol.csv')
```

PM 2.5 column is empty, so drop it and Agency is irrelevant and State is the same. So drop the columns.

```python
df.drop(['Agency','PM 2.5', 'State'], axis = 1, inplace = True)
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Sampling Data'] = pd.to_datetime(df['Sampling Date'],format="%d-%m-%y", dayfirst = True)
```

```python
df.head()
```

```python
# convert date from Sampling Data column to months name, read strftime.
df['Month'] = df['Sampling Data'].dt.strftime('%b')
df.head()
```

Check the mean SO2 and NO2 levels of diff cities throught the year.

```python
x = df.groupby('City/Town/Village/Area')['SO2','NO2'].mean()
x
```

```python
x.plot(kind = 'bar', figsize = (10, 6), title = 'Mean SO2 and NO2 levels')
```

**City Nagaon has higher SO2 and NO2 levels than Guwahati.**

```python
df.groupby('City/Town/Village/Area')['RSPM/PM10'].mean().plot(
    kind = 'bar', figsize = (10,6), title = 'Mean RSPM/PM10 Levels', grid = True)
                                                             
```

RSPM/PM10 is considerably higher in Nagaon than rest of the state. 
Guwahati has less pollution level than many other cities/towns.

Grouping pollution by cities and months.

```python
x = df.groupby(['City/Town/Village/Area', 'Month'])['SO2','NO2','RSPM/PM10'].mean()
x
```

Most and least polluted city/town can be found out by adding three parameters and finding the max and min of them.

```python
x['Total'] = x['SO2'] + x['NO2'] + x['RSPM/PM10']
x.head(15)
```

```python
x.loc[x['Total'].idxmax()]
```

```python
x.loc[x['Total'].idxmin()]
```

Nagaon has highest levels of NO2 and SO2, as well as RSPM/PM10 levels has its pollution level at peak in month of April. April looks like the worst month for the city due to high levels of RSPM/PM10.

Bonaigaon is the cleanest in the month of August.

```python
nagaon = df['City/Town/Village/Area'] == 'Nagaon'
```

```python
nagaon_df = df[nagaon]
nagaon_df.head()
```

```python
ax = sns.barplot(x='Month',y='RSPM/PM10', data = nagaon_df).set_title('Mean RSPM/PM10 level in NAGAON')
```

Highest level of RSPM/PM10 is in months of Apr and May for Nagaon.

```python
sns.barplot(x='Month',y='RSPM/PM10',data=df[df['City/Town/Village/Area']=='Bongaigaon']).set_title('Mean RSPM/PM10 level in Bonaigaon')
```

```python
sns.barplot(x='Month',y='RSPM/PM10',data=df[df['City/Town/Village/Area']=='Guwahati']).set_title('Mean RSPM/PM10 levels in Guwahati')
```

```python
sns.barplot(x='Month',y='RSPM/PM10',data=df[df['City/Town/Village/Area']=='Silcher'],palette='Blues_d').set_title('Mean RSPM/PM10 levels in Silcher')
```

Nagaon, Guwahati and Silcher follow the same trend, i.e high level of RSPM/PM10 in months on march and april.
Bongaigaon being the least polluted town has low levels of RSPM/PM10 even in the months of March and April.
